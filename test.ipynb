```python
#Dropping unnecessary columns
df_songs = df_songs.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'Unnamed: 0'])
df_songs.columns.tolist()
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# ========== 1️⃣ Data Preparation ==========
features = ['valence', 'energy', 'danceability', 'tempo']
X = df_songs[features].dropna()

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ========== 2️⃣ Dimensionality Reduction ==========
# You can tweak n_components for experimentation (2–3 usually works well)
pca = PCA(n_components=2, random_state=42)
X_reduced = pca.fit_transform(X_scaled)

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_.sum():.2%}")

# ========== 3️⃣ KMeans ==========
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_reduced)

# ========== 4️⃣ Gaussian Mixture Model ==========
gmm = GaussianMixture(n_components=4, random_state=42, n_init=5, covariance_type='full')
gmm_labels = gmm.fit_predict(X_reduced)

# ========== 5️⃣ Evaluation ==========
def evaluate(name, labels, X):
    sil = silhouette_score(X, labels)
    dbi = davies_bouldin_score(X, labels)
    chi = calinski_harabasz_score(X, labels)
    print(f"{name} → Silhouette: {sil:.3f} | DBI: {dbi:.3f} | CHI: {chi:.2f}")
    return sil, dbi, chi

print("\n📊 Clustering Evaluation Metrics (on PCA-reduced data):")
kmeans_metrics = evaluate("KMeans", kmeans_labels, X_reduced)
gmm_metrics = evaluate("GMM", gmm_labels, X_reduced)

# ========== 6️⃣ Choose best model automatically ==========
best_labels = gmm_labels if gmm_metrics[0] > kmeans_metrics[0] else kmeans_labels
method_used = "GMM" if gmm_metrics[0] > kmeans_metrics[0] else "KMeans"
df_songs.loc[X.index, 'cluster'] = best_labels

print(f"\n✅ Selected Model: {method_used}")

# ========== 7️⃣ Assign Mood Labels ==========
mood_labels = {
    0: "Happy / Chill",
    1: "Energetic / Party",
    2: "Sad / Mellow",
    3: "Intense / Angry"
}
df_songs.loc[X.index, 'mood_label'] = df_songs.loc[X.index, 'cluster'].map(mood_labels)

# ========== 8️⃣ Visualization ==========
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:, 0], y=X_reduced[:, 1],
                hue=df_songs.loc[X.index, 'mood_label'],
                palette='viridis', alpha=0.7)
plt.title(f'{method_used} Mood Clusters (PCA Reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title="Mood", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Optional: visualize centroids if using KMeans
if method_used == "KMeans":
    centers = kmeans.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')
    plt.legend()
    plt.show()

# ========== 9️⃣ Feature Distribution ==========
sns.pairplot(df_songs.loc[X.index, features + ['mood_label']],
             hue='mood_label', palette='viridis', diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Feature Distributions by Mood Cluster (PCA-reduced)', y=1.02)
plt.show()

# ========== 🔟 Summary ==========
print("\nMood Cluster Averages (original feature scale):")
print(df_songs.groupby('mood_label')[features].mean().round(3))
```

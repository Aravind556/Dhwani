```python
#Dropping unnecessary columns
df_songs = df_songs.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'Unnamed: 0'])
df_songs.columns.tolist()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# features
features = ['valence', 'energy', 'danceability', 'tempo']
X = df_songs[features].dropna()

# Scaling them 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2, random_state=42)
X_reduced = pca.fit_transform(X_scaled)

print(f"Explained variance by PCA components: {pca.explained_variance_ratio_.sum():.2%}")

#Gaussian Mixture Model 
gmm = GaussianMixture(n_components=4, random_state=42, n_init=5, covariance_type='full')
gmm_labels = gmm.fit_predict(X_reduced)

# Add cluster labels to DataFrame
df_songs.loc[X.index, 'cluster'] = gmm_labels

# Evaluation Metrics
silhouette = silhouette_score(X_reduced, gmm_labels)
db_index = davies_bouldin_score(X_reduced, gmm_labels)
ch_index = calinski_harabasz_score(X_reduced, gmm_labels)
#Scores to check 
print("\n📊 GMM Evaluation Metrics (on PCA-reduced data):")
print(f"Silhouette Score: {silhouette:.3f}")
print(f"Davies-Bouldin Index: {db_index:.3f} (lower is better)")
print(f"Calinski-Harabasz Index: {ch_index:.2f} (higher is better)")

# Mood Labeling
mood_labels = {
    0: "Sad / Mellow",
    1: "Energetic / Party",
    2: "Happy / Chill",
    3: "Calm / Content"
}
df_songs.loc[X.index, 'mood_label'] = df_songs.loc[X.index, 'cluster'].map(mood_labels)

# Visualization
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_reduced[:, 0], y=X_reduced[:, 1],
                hue=df_songs.loc[X.index, 'mood_label'],
                palette='viridis', alpha=0.7)
plt.title('🎵 GMM Mood Clusters (PCA Reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title="Mood", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Feature Distributions
sns.pairplot(df_songs.loc[X.index, features + ['mood_label']],
             hue='mood_label', palette='viridis', diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Feature Distributions by Mood Cluster', y=1.02)
plt.show()

# Save Results
output_path = "songs_with_moods_fixed.csv"
df_songs.to_csv(output_path, index=False)
print(f"\n✅ Clustered songs (with mood labels) saved to: {output_path}")



# Example usage:
# get_songs_by_mood("Calm")
```

```python
# ========== 9️⃣ Query Function ==========
def get_songs_by_mood(mood):
    mood = mood.lower()
    results = df_songs[df_songs['mood_label'].str.lower().str.contains(mood)]
    if results.empty:
        print(f"No songs found for mood '{mood}'. Try one of these:")
        print(df_songs['mood_label'].unique())
    else:
        print(f"\n🎶 Songs for mood '{mood}':")
        print(results[['song_name', 'title', 'mood_label']].head(10))  # show sample
    return results
```

```python
get_songs_by_mood("Happy / Chill")
```

```python

def predict_mood(song_features):
    """
    Predict the mood for a song given its features.
    
    song_features: dict with keys as feature names (e.g., {'valence': 0.5, 'energy': 0.8, ...})
    """
    # Convert to DataFrame for consistency
    song_df = pd.DataFrame([song_features])
    
    # Scale using the same scaler
    song_scaled = scaler.transform(song_df)
    
    # Apply PCA using the same pca
    song_reduced = pca.transform(song_scaled)
    
    # Predict cluster with GMM
    cluster = gmm.predict(song_reduced)[0]
    
    # Map to mood label
    mood = mood_labels.get(cluster, "Unknown")
    
    return mood, cluster

#Testing
test_song = {
    'valence': 0.16,  
    'energy': 0.39,
    'danceability': 0.53,
    'tempo': 90.0
}
predicted_mood, predicted_cluster = predict_mood(test_song)
print(f"Predicted Mood: {predicted_mood}")
print(f"Predicted Cluster: {predicted_cluster}")
```
